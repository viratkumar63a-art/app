```python
# Another graph with two lines
from matplotlib import pyplot as plt
from matplotlib import style 
x=[2,3,4]
y=[5,6,7]
x2=[10,11,12]
y2=[13,14,15]
plt.plot(x,y)
plt.xlabel("This is x")
plt.ylabel("this is y")
plt.plot(x2,y2)
style.use('ggplot')
plt.scatter(x,y,label="first",linewidth=3)
plt.plot(x2,y2,label="second")
plt.legend()

plt.show()
```
